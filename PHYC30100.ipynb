### Cross Section/Opacity

#### 1. Assume that the Sun is compsed completely of ionised hydrogen (70%) and helium (30%) and that it is of uniform density.

##### (a) What is the mean free path of a photon in the Sun before it undergoes Thompson scattering? How far into the Sun are we seeing?

The mean free path is the average distance between collisions. The formula is given on slides 101 and 108.

\begin{align}
\lambda &= \frac{1}{\sigma_{\mathrm{T}} \cdot n_{\mathrm{e}}} \\
\end{align}

* $\lambda$, the mean free path, is what we want.
* $\sigma_{\mathrm{T}} = 6.65 \times 10^{-29}$ $\mathrm{m}^{2}$ is the Thompson cross-section for an electron as shown on slide 105.
* $n_{\mathrm{e}}$, the electron number density, is needed.

-----

The formula for electron number density is given on slide 113.

\begin{align}
n_{\mathrm{e}} &= \frac{M}{ \frac{4}{3} \cdot \pi \cdot R^{3} \cdot \mu_{\mathrm{e}} \cdot m_{\mathrm{p}}}
\end{align}

* $n_{\mathrm{e}}$ is what we want.
* $M = 1.989 \times 10^{30}$ $\mathrm{kg}$ is the solar mass.
* $\pi = 3.1415 $ is pi.
* $R = 6.957 \times 10^{8}$ $\mathrm{m}$ is the solar radius.
* $m_{\mathrm{p}} = 1.7 \times 10^{-27}$ $\mathrm{kg}$ is the mass of the proton.
* $\mu_{\mathrm{e}}$, the average number of free electrons per nucleon, is needed.

---

There is a formula for the average number of free electrons per nucleon.

\begin{align*}
\frac{1}{{\mu}_\mathrm{e}} &= \sum \frac{X_{i} \cdot Z_{i}}{A_{i}} \\
                 &= \frac{X_{\mathrm{H}} \cdot Z_{\mathrm{H}}}{A_{\mathrm{H}}} + \frac{X_{\mathrm{He}} \cdot Z_{\mathrm{He}}}{A_{\mathrm{He}}}
\end{align*}

* $\mu_{\mathrm{e}}$ is what we want.
* $X_{\mathrm{H}} = 0.7$ is the relative fraction of ionised hydrogen.
* $Z_{\mathrm{H}} = 1$ $\mathrm{amu}$ is the atomic number of ionised hydrogen.
* $A_{\mathrm{H}} = 1$ is the mass number of ionised hydrogen.
* $X_{\mathrm{He}} = 0.3$ the relative fraction of ionosed helium.
* $Z_{\mathrm{He}} = 2$ $\mathrm{amu}$ is the atomic number of ionised helium.
* $A_{\mathrm{He}} = 4$ is the mass number of ionised helium.

\begin{align*}
\frac{1}{{\mu}_\mathrm{e}}&= \frac{(0.7) \cdot (1)}{(1)} + \frac{(0.3) \cdot (2)}{(4)} \\
&= 0.85 \\
\mu_{\mathrm{e}} &= \frac{1}{0.85} \\
&\approx 1.18
\end{align*}

The average number of free electrons per nucleon is $1.18$. 

-----

Plug $\mu_{\mathrm{e}} = 1.18$ into the equation $2$.

\begin{align*}
n_{\mathrm{e}} &= \frac{M}{ \frac{4}{3} \cdot \pi \cdot R^{3} \cdot \mu_{\mathrm{e}} \cdot m_{\mathrm{p}}} \\
&= \frac{(1.989 \times 10^{30})}{ \frac{4}{3} \cdot (3.1415) \cdot (6.957 \times 10^{8})^{3} \cdot (1.18) \cdot (1.7 \times 10^{-27})} \\ 
&\approx 7 \times 10^{29} \; \mathrm{m}^{-3}
\end{align*}

-----

Plug $n_{\mathrm{e}} = 7 \times 10^{29}$ $\mathrm{m}^{-3}$ into the equation $1$.

\begin{align*}
\lambda &= \frac{1}{\sigma_{\mathrm{T}} \cdot n_{\mathrm{e}}} \\
&= \frac{1}{(6.65 \times 10^{-29}) \cdot (7 \times 10^{29})} \\
&\approx 0.021 \; \mathrm{m}
\end{align*}

The mean free path of a photon in the Sun before it undergoes Thompson scattering is $2.1$ $\mathrm{cm}$.

-----

The formula for the attenuation of flux is given on slide 101.

\begin{align*}
F = F_{0} \cdot \mathrm{e}^{-\sigma \cdot n \cdot x}
\end{align*}

When dealing with light, $\sigma \cdot n \cdot x$ is known as the optical depth, as stated on slide 101. When the optical depth is $1$, the medium is opaque. This represents how far into the Sun we are seeing.

\begin{align*}
\sigma \cdot n \cdot x &= 1 \\
x &= \frac{1}{\sigma \cdot n} \\
&= \lambda \\
&= 0.021 \; \mathrm{m}
\end{align*}

It is clear from equation $1$ that $x = \lambda$ so the answer is the same. We see to a depth of $2.1$ $\mathrm{cm}$.

-----

##### (b) How many scatterings will a photon produced at the centre undergo before it reaches the surface?

The formula for the number of times a photon has been scattered is on slide 111.

\begin{align*}
n &= \bigg(\frac{R}{\lambda}\bigg)^{2} \\
\end{align*}

* $n$, the number of times a photon has scattered, is what we want.
* $R = 6.957 \times 10^{8}$ $\mathrm{m}$ is the solar radius.
* $\lambda = 0.021$ $\mathrm{m}$ is the mean free path.

\begin{align*}
n &= \bigg(\frac{(6.957 \times 10^{8})}{(0.021)}\bigg)^{2} \\
&\approx 1.1 \times 10^{21}
\end{align*}

-----

##### (c) What total distance will it have traveled to reach the surface?

The distance traveled is the distance the photon travels when it is scattered multiplied by the amount of times it has been scattered.

\begin{align*}
D &= n \cdot \lambda
\end{align*}

* $D$, the distance, is what we want.
* $n = 1.1 \times 10^{21}$ is the number of scatterings a photon will undergo.
* $\lambda = 0.021$ $\mathrm{m}$ is the mean free path.

\begin{align*}
D &= (1.1 \times 10^{21}) \cdot (0.021) \\
&= 2.3 \times 10^{19} \; \mathrm{m}
\end{align*}

A photon will have traveled $2.3 \times 10^{19}$ $\mathrm{m}$ by the time it reaches the surface.

-----

##### (d) How long will it take the photon to reach the surface?

Velocity is distance divided by time. Therefore, the time taken for a photon to reach the surface is the distance divided by the speed of light.

\begin{align*}
t = \frac{D}{c}
\end{align*}

* $t$, the time taken to reach the surface, is what we want.
* $D = 2.3 \times 10^{19}$ $\mathrm{m}$, is the distance the photon travels.
* $c = 3 \times 10^{8}$ $\mathrm{m} \cdot \mathrm{s}^{-1}$ is the speed of light.

\begin{align*}
t &= \frac{(2.3 \times 10^{19})}{(3 \times 10^{8})} \\
&= 7.7 \times 10^{10} \; \mathrm{s}
\end{align*}

It takes the photon $7.7 \times 10^{10}$ $\mathrm{s}$ = $2\,440$ $\mathrm{yr}$ to reach the surface.

#### 2. The cross-section for the interaction of a $1$ $\mathrm{GeV}$ photon with lead is $40$ $\mathrm{barn} \cdot \mathrm{atom}^{-1}$. What thickness of lead shielding would be necessary to block $99\%$ of the incident flux of $1$ $\mathrm{GeV}$? Lead has a density of $11\,350$ $\mathrm{kg} \cdot \mathrm{m}^{-3}$ and atomic mass of $207.2$ $\mathrm{amu}$.

The formula for the attenuation of flux is given on slide 101.

\begin{align}
F = F_{0} \cdot \mathrm{e}^{-\sigma \cdot n \cdot x} \\
\end{align}

* $x$, the thickness, is what we want.
* $F = 100\% - 99\% = 1\% = 0.01$ is the transmitted intensity.
* $F_{0} = 100\% = 1$, the incident intensity.
* $\sigma = 40$ $\mathrm{b} \cdot \mathrm{atom}^{-1} = 40 \times 10^{-28}$ $\mathrm{m}$ is the cross-section for the interaction.
* $n$, the particle number density, is needed.

Rearrange equation $3$ to solve for $x$.

\begin{align}
F &= F_{0} \cdot \mathrm{e}^{-\sigma \cdot n \cdot x} \nonumber \\
\frac{F}{F_0} &= \mathrm{e}^{-\sigma \cdot n \cdot x} \nonumber \\
\ln \bigg(\frac{F}{F_0}\bigg) &= -\sigma \cdot n \cdot x \nonumber \\
\frac{\ln \big(\frac{F}{F_0}\big)}{\sigma \cdot n} &= -x \nonumber \\
x &= \frac{-\ln \big(\frac{F}{F_0}\big)}{\sigma \cdot n}
\end{align}

-----
There is a formula to find the particle number density.

\begin{align*}
n &= \frac{\rho}{A \cdot m_{\mathrm{p}}}
\end{align*}

* $n$ is what we want.
* $\rho = 11\,350$ $\mathrm{kg} \cdot \mathrm{m}^{-3}$ is the density.
* $A = 207.2$ $\mathrm{amu}$ is the atomic mass.
* $m_\mathrm{p} = 1.7 \times 10^{-27}$ $\mathrm{kg}$ is the proton mass.

\begin{align*}
n &= \frac{(11\,350)}{(207.2) \cdot (1.7 \times 10^{-27})} \\
&= 3.2 \times 10^{28} \; \mathrm{m}^{-3}
\end{align*}

The particle number density is $3.2 \times 10^{28}$ $\mathrm{m}^{-3}$. 

-----

Plug $n = 3.2 \times 10^{28}$ $\mathrm{m}^{-3}$ into equation $4$.

\begin{align*}
x &= \frac{-\ln \big(\frac{F}{F_0}\big)}{\sigma \cdot n} \\
&= \frac{-\ln \big(\frac{(0.01)}{(1)}\big)}{(40 \times 10^{-28}) \cdot (3.2 \times 10^{28})} \\
&= 0.036 \; \mathrm{m}
\end{align*}

The lead shielding would need to be $3.6$ $\mathrm{cm}$ thick.
